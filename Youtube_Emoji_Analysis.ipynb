```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import emoji
```

```python
df = pd.read_csv('C:/Users/shishgup/Desktop/Hangman-DataFlair/Youtube_Text_Data_Analysis/GBcomments.csv', error_bad_lines=False)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
str = ''
for i in df['comment_text']:
    lst = [c for c in i if c in emoji.UNICODE_EMOJI_ENGLISH]
    for element in lst:
        str = str + element
```

```python
str
```

```python
dctnry = {}
for j in set(str):
    dctnry[j] = str.count(j)
```

```python
dctnry
```

```python
final_dct = {}
for k,v in sorted(dctnry.items(), key = lambda item: item[1]):
    final_dct[k] = v
final_dct
```

```python
keys = [*final_dct.keys()]
```

```python
keys
```

```python
values = [*final_dct.values()]
values
```

```python
new_df = pd.DataFrame({'chars':keys[-20:], 'nums':values[-20]})
```
